```python
timestamps = []
with open(os.path.join("test_data", "splitted", "timestamps.csv"), newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csvreader:
        timestamps.append(datetime.datetime.fromisoformat(*row))
        
dt = (timestamps[-1] - timestamps[0]).total_seconds()

hours, remainder = divmod(dt, 3600)
minutes, seconds = divmod(remainder, 60)
hours, minutes, seconds
print("{:02d}:{:02d}:{:0.3f}".format(int(hours), int(minutes), seconds))
```
